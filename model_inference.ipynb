# Model Inference

```python
# Import Libaries
import pickle
import json
import pandas as pd
import numpy as np
```

# Load Model
> Load model yang telah dibuat pada notebook berbeda

```python
# Load all models dari perhitungan sebelumnya

with open('model.pkl', 'rb') as  file_1: 
  model = pickle.load(file_1)

with open('scaler.pkl', 'rb') as file_2:
  scaler = pickle.load(file_2)

with open('pca.pkl', 'rb') as file_3:
  pca = pickle.load(file_3)

with open('num_col.txt', 'r') as file_4:
  num_col = json.load(file_4)

with open('cat_col.txt', 'r') as file_5:
  cat_col = json.load(file_5)
```

# Create Data Inference
> Membuat data baru untuk mencoba model yang telah dibuat

Menginput value dari kolom-kolom pada data_inf

```python
# create new data
data_inf = {
    'country': ['Canada', 'Cape Verde', 'Central African Republic', 'Chad', 'Chile', 'China', 'Colombia', 'Comoros'],
    'child_mortality': [5.6, 26.5, 149.0, 150.0, 8.7, 15.7, 18.6, 88.2],
    'exports': [29.1, 32.7, 11.8, 36.8, 37.7, 26.3, 15.9, 16.5],
    'health': [11.3, 4.09, 3.98, 4.53, 7.96, 5.07, 7.59, 4.51],
    'imports': [31.0, 61.8, 26.5, 43.5, 31.3, 22.6, 17.8, 51.7],
    'income': [40700, 5830, 888, 1930, 19400, 9530, 10900, 1410],
    'inflation': [2.87, 0.505, 2.01, 6.39, 8.96, 6.94, 3.86, 3.87],
    'life_expectancy': [81.3, 72.5, 47.5, 56.5, 79.1, 74.6, 76.4, 65.9],
    'total_fertility': [1.63, 2.67, 5.21, 6.59, 1.88, 1.59, 2.01, 4.75],
    'gdpp': [47400, 3310, 446, 897, 12900, 4560, 6250, 769]
}

data_inf = pd.DataFrame(data_inf)
data_inf
```

# Langkah Hingga Prediksi

```python
# Split between numerical and categorical columns
data_inf_num = data_inf[num_col]
data_inf_cat = data_inf[cat_col]

# melihat kolom kategorikal
data_inf_cat
```

```python
# melihat kolom numerikal
data_inf_num
```

```python
# Feature scaling pada numerik
data_inf_num_scaled = scaler.transform(data_inf_num)
```

```python
# pca
data_num_pca = pca.fit(data_inf_num_scaled)
df_scaled_pca = pca.transform(data_inf_num_scaled)
```

```python
# membuat dataframe dengan PCA
df_scaled_pca = pd.DataFrame(df_scaled_pca, columns=['PCA1', 'PCA2', 'PCA3', 'PCA4', 'PCA5', 'PCA6', 'PCA7', 'PCA8'])
```

```python
# Concate 
data_inf_final = np.concatenate([df_scaled_pca, data_inf_cat],axis=1)
```

Membuat predict dengan model k-means yang telah dibuat

```python
# Predict new data inference using Linear Regression
pred = model.fit_predict(df_scaled_pca)
pred
```

```python
# insert kolom cluster pada dataframe asli
data_inf['cluster'] = pred

# menampilkan dataframe
data_inf
```

Menjadi 11 kolom

# Melihat Karakteristik Clustering

```python
# melihat berdasarkan kematian anak
data_inf.groupby(['cluster'])[['child_mortality']].mean()
```

Rata-rata kematian anak tertinggi pada cluster 1

Rata-rata pendapatan terendah pada cluster 1
